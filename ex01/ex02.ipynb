#### tensorflow indexing 

```python
import numpy as np 
import tensorflow as tf 
```

```python
vec = tf.constant([10, 20, 30, 40, 50])
print(vec)
print(vec[-1])
print(vec[:3])
```

```python
mat = tf.constant([[10, 20, 30],[40, 50, 60]])
print(mat)
print(mat[0, 2])
print(mat[0, :])
print(mat[:, 1])
```

```python
ten = tf.constant([
                   [[10, 20, 30],
                    [40, 50, 60]], 
                  [[-10, -20, -30], 
                   [40, 50, 60]]])
print(ten)
print(ten[0, :, :])
print(ten[:, :2, :2])
```
