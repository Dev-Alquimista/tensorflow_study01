#### tensorflow reshape 

```python
import tensorflow as tf 
import numpy as np 
```

```python
ten1 = tf.constant(range(0, 24))
print(ten1)

ten2 = tf.reshape(ten1, [3, 8])
print(ten2)
```

```python
ten3 = tf.reshape(ten1, [-1, 4])
print(ten3)
```

```python
ten4 = tf.reshape(ten3, [-1])
print(ten4)
```

```python
ten5 = tf.reshape(ten4, [-1, 3, 4])
print(ten5)
```
