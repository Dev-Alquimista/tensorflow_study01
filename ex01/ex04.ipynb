```python
import numpy as np 
import tensorflow as tf 
```

tensorflow variable 



```python
ten1 = tf.constant([[0, 1, 2],
                  [3, 4, 5]])
print(ten1)
```

```python
ten_var1 = tf.Variable(ten1)
print(ten_var1)
```

```python
print(f"Name : {ten_var1.name}")
print(f"Shape : {ten_var1.shape}")
print(f"data type : {ten_var1.dtype}")
print(f"array : {ten_var1.numpy()}")
```

```python
ten_var1.assign([[1, 1, 1],
                 [2, 2, 2]])
```

```python
print(f"Name : {ten_var1.name}")
print(f"Shape : {ten_var1.shape}")
print(f"data type : {ten_var1.dtype}")
print(f"array : {ten_var1.numpy()}")
```
