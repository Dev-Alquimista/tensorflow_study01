```python
import numpy as np 
import tensorflow as tf 
import matplotlib.pyplot as plt 
```

#### gradient descent 
$$\frac{\partial{Loss}}{\partial{w}} = \frac{\partial}{\partial{w}}(\sum_1y-\hat{y})^2 $$
$$=\sum_1(y-\hat{y})\frac{âˆ‚}{patial}$$

```python
def set_make_linear(w=0.5, b=0.8, size=50, noise=1.0):
    x = np.random.rand(size)
    y = w * x + b  
    noise = np.random.uniform(-abs(noise), abs(noise), size=y.shape)
    print(noise.shape)    
    yy = y + noise 
    get_show_graph(w, x, y, yy, b)

    print(f'w: {w}, b: {b}')
    return x, yy

def get_show_graph(w, x, y, yy, b):
    plt.figure(figsize=(22, 7))
    plt.plot(x, y, color="r", label=f"y={w}*x + {b}")
    plt.scatter(x, yy, label="data")
    plt.legend(fontsize=20)
    plt.show() 
   
```

```python
def get_optimizer_value(x, y):
    num_epoch=1000
    learning_rate = 0.005

    errors = []

    w = np.random.uniform(low=0.0, high=1.0)
    b = np.random.uniform(low=0.0, high=1.0)

    for epoch in range(num_epoch):

        y_hat = w * x + b 
        error = 0.5*((y_hat - y) ** 2).sum()

        if error < 0.005:
            break 

        w = w - learning_rate * ((y_hat - y)*x).sum()
        b = b - learning_rate * (y_hat - y).sum()


        if epoch % 5 == 0 :
            print("{0:2} w = {1:.5f}, b={2:.5f} error = {3:.5f}".format(epoch, w, b, error))
        print("-----"*15)
        print("{0:2} w = {1:.5f}, b={2:.5f} error = {3:.5f}".format(epoch, w, b, error))

        errors.append(error)
    return errors 
```

```python
def get_show_error(errors):
    plt.figure(figsize=(10, 7))
    plt.plot(errors)
    plt.xlabel('Epochs')
    plt.ylabel('Error')
    plt.show()
```

```python
if __name__ == "__main__":
    x, y = set_make_linear(w=0.3, b=0.5, size=100, noise=0.01)
    errors = get_optimizer_value(x, y)
    get_show_error(errors)
```
