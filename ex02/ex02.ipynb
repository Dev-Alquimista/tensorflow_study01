```python
import cv2
import numpy as np 
import pandas as pd 
import tensorflow as tf 
import matplotlib.pyplot as plt 
```

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense 
from tensorflow.keras.optimizers import Adam, RMSprop, SGD 
from tensorflow.keras.metrics import mean_absolute_error, mean_squared_error
```

```python
def get_show_graph(x, y):
    plt.plot(x, y, label="f(x) = 3x + 2")
    plt.grid(True)
    plt.axis("on")
    plt.scatter(x, y, label="data", color="r")
    plt.legend()
```

```python
x = np.arange(1, 6)
print(type(x))
print(x.shape)
print(x)

y = 3 * x + 2 
print(type(y))
print(y.shape)
print(y)

get_show_graph(x, y)
```

```python
model = Sequential([tf.keras.layers.Dense(1, input_shape=[1])])
model.summary()
```

```python
# model.compile(optimizer="sgd", 
#               loss="mean_squared_error", 
#               metrice=["mean_squared_error", "mean_absolute_error"])

# model.compile(optimizer="sgd", 
#               loss="mse", 
#               metrics=["mse", "mae"])

# model.compile(optimizer=SGD(lr=0.005),
#               loss=[mean_absolute_error, mean_squared_error]) 
```

```python
model.compile(optimizer="sgd", loss="mse", metrics=["mae"])
history = model.fit(x, y, epochs=1200)
```

```python
plt.plot(history.history["loss"], label="loss")
plt.plot(history.history["mae"], label="mae")
plt.xlim(-1, 20)
plt.title("Loss")
plt.legend()
plt.show()
```

```python
model.evaluate(x, y)
model.predict([10])
```
